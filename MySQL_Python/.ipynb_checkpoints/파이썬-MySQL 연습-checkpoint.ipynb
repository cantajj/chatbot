### TABLE이 이미 존재한다면 DROP|

```python
db = None
try:
    # DB 호스트 정보에 맞게 입력해주세요
    db = pymysql.connect(
        host='127.0.0.1',
        user='root',
        passwd='1234',
        db='test_db',
        charset='utf8'
    )
    print("DB 연결 성공")
    
    # 테이블 생성 sql 정의
    sql = '''
            DROP TABLE IF EXISTS tb_student
    '''
    # 테이블 생성
    with db.cursor() as cursor:
        cursor.execute(sql)
    print("TABLE DROP 성공")
    
except Exception as e:
    print(e)

finally:
    if db is not None:
        db.close()
        print("DB 연결 닫기 성공")
```

### 테이블 생성
* auto_increment : Insert할 때 마다 숫자가 자동으로 증가하며 부여됨

```python
db = None
try:
    # DB 호스트 정보에 맞게 입력해주세요
    db = pymysql.connect(
        host='127.0.0.1',
        user='root',
        passwd='1234',
        db='test_db',
        charset='utf8'
    )
    print("DB 연결 성공")
    
    # 테이블 생성 sql 정의
    sql = '''
            CREATE TABLE tb_student (
                id int primary key auto_increment not null primary key,
                name varchar(32),
                age int,
                address varchar(32)
            ) ENGINE=InnoDB DEFAULT CHARSET=utf8
    '''
    # 테이블 생성
    with db.cursor() as cursor:
        cursor.execute(sql)
    print("TABLE CREATE 성공")
    
except Exception as e:
    print(e)

finally:
    if db is not None:
        db.close()
        print("DB 연결 닫기 성공")
```

### 다량 데이터 INSERT

```python
# 데이터 정의
students = [
    {'name': '김남준', 'age': 28, 'address' : '서울'},
    {'name': '김석진', 'age': 30, 'address': '과천'},
    {'name': '민윤기', 'age': 28, 'address': '대구'},
    {'name': '전정국', 'age': 25, 'address': '부산'},
    {'name': '김태형', 'age': 26, 'address': '대구'},
    {'name': '정호석', 'age': 28, 'address': '광주'},
    {'name': '박지민', 'age': 27, 'address': '부산'},
]

# 데이터 DB에 INSERT
for s in students:
    with db.cursor() as cursor:
        sql = '''
            INSERT tb_student(name, age, address)
            VALUES('%s', %d, '%s')        
        ''' % (s['name'], s['age'], s['address'])
        cursor.execute(sql)
        
db.commit()
```

### fetchall() 로 모두 가져오기

```python
### 딕셔너리 커서로 가져오기 (json형식)
with db.cursor(pymysql.cursors.DictCursor) as cursor:
    sql = '''
        SELECT * 
        FROM tb_student
    '''
    cursor.execute(sql)
    result = cursor.fetchall()  # 데이터 모두 가져오기
result
```

### try except 안에 써보기

```python
import pymysql
import pandas as pd

try:
    db = pymysql.connect(
            host='127.0.0.1',
            user='root',
            passwd='1234',
            db='test_db',
            charset='utf8'
    )
    # DB 호스트 정보에 맞게 입력해주세요
    db = pymysql.connect(
        host='127.0.0.1',
        user='root',
        passwd='1234',
        db='test_db',
        charset='utf8'
    )
    
    # 데이터 정의
    students = [
        {'name': '김남준', 'age': 28, 'address' : '서울'},
        {'name': '김석진', 'age': 30, 'address': '과천'},
        {'name': '민윤기', 'age': 28, 'address': '대구'},
        {'name': '전정국', 'age': 25, 'address': '부산'},
        {'name': '김태형', 'age': 26, 'address': '대구'},
        {'name': '정호석', 'age': 28, 'address': '광주'},
        {'name': '박지민', 'age': 27, 'address': '부산'},
    ]

    # 데이터 db에 추가
    for s in students:
        with db.cursor() as cursor:
            sql = '''
                    insert tb_student(name, age, address) 
                    values("%s",%d,"%s")
                    ''' % (s['name'], s['age'], s['address'])
            cursor.execute(sql)
    db.commit() # 커밋    
    
    
    # 28세 이상 멤버만 조회
    cond_age = 27
    with db.cursor(pymysql.cursors.DictCursor) as cursor:
        sql = ''' 
        select * from tb_student where age > %d
        ''' % cond_age
        cursor.execute(sql)
        results = cursor.fetchall()
    print(results)

    # 이름 검색
    cond_name = '정호석'
    with db.cursor(pymysql.cursors.DictCursor) as cursor:
        sql = ''' 
        select * from tb_student where name="%s"
        ''' % cond_name
        cursor.execute(sql)
        result = cursor.fetchone()
    print(result['name'], result['age'])

    # pandas 데이터프레임으로 표현
    df = pd.DataFrame(results)
    print(df) 
    
except Exception as e:
    print(e)

finally:
    if db is not None:
        db.close()
```
