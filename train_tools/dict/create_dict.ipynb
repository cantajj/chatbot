### 전체코드

```python
##############################################################
# <실습 8> 단어사전 생성
# /train_tools/dict/create_dict.ipynb
# 챗봇에서 사용하는 사전 파일 생성
# corpus.txt를 읽어서 단어사전 chatbot_dict.bin 파일 생성
##############################################################
import sys
sys.path.append('C:/chatbot')

from utils.Preprocess import Preprocess
from tensorflow.keras import preprocessing
import pickle

# 말뭉치 데이터 읽어오기
def read_corpus_data(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        data = data[1:] # 헤더 제거
    return data

# 말뭉치 데이터 가져오기
corpus_data = read_corpus_data('./corpus.txt')

# 말뭉치 데이터에서 키워드만 추출해서 사전 리스트 생성
p = Preprocess()  # 사용자 사전 없음 (default : userdic=None)
dict = []
for c in corpus_data:      # c : 리스트로 만들어진 한 줄 ['0000', ' 한  문장 ', '0/1/2']
    pos = p.pos(c[1])      # pos함수에 문장을 넣어 품사 분석
    for k in pos:          # pos: [(단어, 품사), (단어, 품사), ...)]
        dict.append(k[0])  # 단어만 사전에 추가

# 사전에 사용될 word2index 생성
# 사전의 없는 첫번째 인덱스에는 OOV 사용 (Out Of Vocabulary word)
tokenizer = preprocessing.text.Tokenizer(oov_token='OOV')
tokenizer.fit_on_texts(dict)
word_index = tokenizer.word_index

# 사전 파일 생성
f = open("chatbot_dict.bin", "wb")
try:
    pickle.dump(word_index, f)
    print("단어 사전 chatbot_dict.bin 파일 생성 성공!!!")
except Exception as e:
    print(e)
finally:
    f.close()
```
