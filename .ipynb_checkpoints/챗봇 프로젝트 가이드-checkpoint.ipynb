## 4. 데이터베이스 제어 모듈 생성

### <실습 18> 데이터베이스 제어 모듈 생성 (class Database)
* 데이터베이스 제어 모듈 (class Database) : `/utils/Database.py`
* 함수 정의
    - close() : DB연결 닫기
    - execute(sql) : SQL 구문 실행
    - select_one(sql) : 단 1개의 데이터 ROW만 불러옴
    - select_all(sql) : 전체 데이터 ROW를 불러옴

## 5. 답변 검색 및 답변 출력

### <실습 19> 챗봇 답변 검색 모듈  생성 : (class FindAnswer)
* 챗봇 답변 검색 모듈 : `/utlis/FindAnswer.py` (class FindAnswer)
* 내장 함수(method)
    - \_\_init__(db) : db를 입력받음
    - \_make_query(intenet_name, ner_tags) : intent와 ner을 입력받아 sql문 작성
    - search(intent_name, ner_tags) : intent와 ner을 입력받아 답변 검색하여 답변과 답변 이미지 return
    - tag_to_word(ner_predicts, answer) : NER 태그를 실제 입력된 단어로 변환
